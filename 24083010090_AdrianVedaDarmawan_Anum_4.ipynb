Mencari luas dalam pagar terbesar dengan jumlah pagar terbatas. Asumsikan jumlah pagarnya 64

```python
fences = 64
def make_farm(l, w):
    cost = 2 * (l + w) + 4
    if cost > fences:
        return "Not enough fencing"
    else:
        area = max(l - 2, 0) * max(w - 2, 0)
        return {
            "area": area,
            "cost": cost
        }

for i in range(0, 30+1):
    result = make_farm(i, 30-i)
    print(f"W: {i}, L: {30-i}, Fence Cost: {result["cost"]}, Farm area: {result["area"]} blocks")
```

```python
max_area = 0
dimension_for_max_area = (0, 0)
cost_for_max_area = 0
for i in range(0, 50+1): #coba angka yang melebihi batas
    for j in range(0, 50+1): #coba angka yang melebihi batas
        result = make_farm(i, j)
        if result not in ["Not enough fencing", None]: #angka yang melebihi batas jumlah pagar dibuang
            if result["area"] > max_area:
                max_area = result["area"]
                cost_for_max_area = result["cost"]
                dimension_for_max_area = (i, j)
print(f"Max area: {max_area} blocks (for size {dimension_for_max_area} and cost of {cost_for_max_area})")
```
